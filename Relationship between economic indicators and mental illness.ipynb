```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import matplotlib.dates as mdates
import numpy as np
from scipy.stats import pearsonr
```

```python
mi_file_path = 'data_mental_illness.csv'
mi_data = pd.read_csv(mi_file_path)
```

```python
plt.figure(figsize=(20, 10))
```

```python
plt.plot(mi_data['year'], mi_data['total'], marker='o', label='Total', color='navy')

plt.ylim(0, 0.35)
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(1.0))

plt.legend()
plt.grid(False)
plt.show()
```

```python
plt.plot(mi_data['year'], mi_data['ages_18_25'], marker='o', label='Ages 18-25', color='mediumseagreen')
plt.plot(mi_data['year'], mi_data['ages_26_49'], marker='o', label='Ages 26-49', color='mediumorchid')
plt.plot(mi_data['year'], mi_data['ages_50_or_Older'], marker='o', label='Ages 50 or Older', color='peru')

plt.ylim(0, 0.35)
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(1.0))

plt.legend()
plt.grid(False)
plt.show()
```

```python
file_path_gini_mental = 'gini_mental_illness.csv'
data_gini_mental = pd.read_csv(file_path_gini_mental)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data_gini_mental['gini_racio'], data_gini_mental['mental_illness'], color='navy')

m, b = np.polyfit(data_gini_mental['gini_racio'], data_gini_mental['mental_illness'], 1)
plt.plot(data_gini_mental['gini_racio'], m*data_gini_mental['gini_racio'] + b, color='red', label='Trend line')

plt.grid(False)
plt.show()
```

```python
# Calculate the Pearson correlation coefficient and the p-value
correlation, p_value = pearsonr(data_gini_mental['mental_illness'], data_gini_mental['gini_racio'])

correlation, p_value
```
